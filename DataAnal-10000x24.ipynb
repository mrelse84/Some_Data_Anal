```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import text

import sys
print('Python Ver. : ', sys.version)
```

## 데이터 로드

```python
data = pd.read_csv("./TestResults-4-10000x24.csv", delimiter=",", header=0) # 컬럼영으로 사용할 행의 번호를 0으로 설정
```

```python
data
```

```python
data_cvt = data*4.4 # pixel to um
```

```python
data_cvt
```

```python
data_cvt.describe()
```

```python
# Shift center to zero
data_cvt_shift = data_cvt.copy()
rows = data_cvt.shape[0]
cols = data_cvt.shape[1]
for icol in range(0, cols):
    data_cvt_shift.iloc[:,icol] = data_cvt.iloc[:,icol] - data_cvt.iloc[:,icol].mean()
```

```python
data_cvt_shift.describe()
```

```python
plt.figure(figsize=(12, 8))
plt.title("Position X[um]")
x_cols = ['P1-X', 'P2-X', 'P3-X', 'P4-X', 'P5-X', 'P6-X', 'P7-X', 'P8-X', 'P9-X', 'P10-X', 'P11-X', 'P12-X']
data_cvt.boxplot(column=x_cols)
```

```python
plt.figure(figsize=(12, 8))
plt.title("Position Y [um]")
x_cols = ['P1-Y', 'P2-Y', 'P3-Y', 'P4-Y', 'P5-Y', 'P6-Y', 'P7-Y', 'P8-Y', 'P9-Y', 'P10-Y', 'P11-Y', 'P12-Y']
data_cvt.boxplot(column=x_cols)
```

```python
plt.figure(figsize=(12, 8))
plt.title("Shifted Position X [um]")
x_cols = ['P1-X', 'P2-X', 'P3-X', 'P4-X', 'P5-X', 'P6-X', 'P7-X', 'P8-X', 'P9-X', 'P10-X', 'P11-X', 'P12-X']
data_cvt_shift.boxplot(column=x_cols)
```

```python
plt.figure(figsize=(12, 8))
plt.title("Shifted Position Y [um]")
x_cols = ['P1-Y', 'P2-Y', 'P3-Y', 'P4-Y', 'P5-Y', 'P6-Y', 'P7-Y', 'P8-Y', 'P9-Y', 'P10-Y', 'P11-Y', 'P12-Y']
data_cvt_shift.boxplot(column=x_cols)
```
