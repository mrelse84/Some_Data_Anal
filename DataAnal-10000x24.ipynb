# Data Analysis of 10000 by 24

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import text

import sys
print('Python Ver. : ', sys.version)
```

## 데이터 로드

```python
data = pd.read_csv("./TestResults-4-10000x24.csv", delimiter=",", header=0) # 컬럼명으로 사용할 행의 번호를 0으로 설정
```

```python
data
```

```python
data_cvt = data*4.4 # pixel to um
```

```python
data_cvt
```

```python
data_cvt.describe()
```

```python
# Shift center to zero
data_cvt_shift = data_cvt.copy()
rows = data_cvt.shape[0]
cols = data_cvt.shape[1]
for icol in range(0, cols):
    data_cvt_shift.iloc[:,icol] = data_cvt.iloc[:,icol] - data_cvt.iloc[:,icol].mean()
```

```python
data_cvt_shift.describe()
```

```python
plt.figure(figsize=(12, 8))
plt.title("Position X[um]")
x_cols = ['P1-X', 'P2-X', 'P3-X', 'P4-X', 'P5-X', 'P6-X', 'P7-X', 'P8-X', 'P9-X', 'P10-X', 'P11-X', 'P12-X']
data_cvt.boxplot(column=x_cols)
```

```python
plt.figure(figsize=(12, 8))
plt.title("Position Y [um]")
x_cols = ['P1-Y', 'P2-Y', 'P3-Y', 'P4-Y', 'P5-Y', 'P6-Y', 'P7-Y', 'P8-Y', 'P9-Y', 'P10-Y', 'P11-Y', 'P12-Y']
data_cvt.boxplot(column=x_cols)
```

```python
plt.figure(figsize=(12, 8))
plt.title("Shifted Position X [um]")
x_cols = ['P1-X', 'P2-X', 'P3-X', 'P4-X', 'P5-X', 'P6-X', 'P7-X', 'P8-X', 'P9-X', 'P10-X', 'P11-X', 'P12-X']
data_cvt_shift.boxplot(column=x_cols)
```

```python
plt.figure(figsize=(12, 8))
plt.title("Shifted Position Y [um]")
x_cols = ['P1-Y', 'P2-Y', 'P3-Y', 'P4-Y', 'P5-Y', 'P6-Y', 'P7-Y', 'P8-Y', 'P9-Y', 'P10-Y', 'P11-Y', 'P12-Y']
data_cvt_shift.boxplot(column=x_cols)
```

```python
# Radius 계산
data_cvt_shift_r = data_cvt_shift.copy()
rows = data_cvt_shift_r.shape[0]
cols = data_cvt_shift_r.shape[1]
for idx in range(0, 12):
    data_cvt_shift_r['P{}-R'.format(idx+1)] = np.sqrt(data_cvt_shift_r[data_cvt_shift_r.columns[idx*2]]**2+data_cvt_shift_r[data_cvt_shift_r.columns[idx*2+1]]**2)
```

```python
fig = plt.figure(figsize=(16,50), facecolor="w")
#fig.tight_layout()
fig.subplots_adjust(hspace=0.15)
fig.subplots_adjust(wspace=0.1)


for i in range(0, 12):
    feature_name = data_cvt_shift_r.columns[24+i]
    avg=data_cvt_shift_r[feature_name].mean()
    sigma=data_cvt_shift_r[feature_name].std()
    
    plt.subplot(6, 2, i+1)
    plt.title(r"Scatter Plot of Center-shifted Picker-{} Position".format(i+1))
    
    # 산점도 그리기
    plt.scatter(data_cvt_shift[data_cvt_shift.columns[i*2]], data_cvt_shift[data_cvt_shift.columns[i*2+1]], marker='.')
    
   # 좌표축 그리기
    plt.axhline(y=0, color='b', linestyle='-', linewidth=1)
    plt.axvline(x=0, color='b', linestyle='-', linewidth=1) 
    
    # 1*sigma 원 그리기
    circle_1s = plt.Circle((0.0, 0.0), sigma, fill=False, color='r', linewidth=1, linestyle=':')
    plt.gca().add_artist(circle_1s)
    text(10, sigma, r"$\bar{r}+$" + r"$\sigma : {}\ [um]$".format(round(avg+sigma,2)), rotation=0, verticalalignment='center')

    # 2*sigma 원 그리기
    circle_2s = plt.Circle((0.0, 0.0), sigma*2, fill=False, color='r', linewidth=1, linestyle='-.')
    plt.gca().add_artist(circle_2s)
    text(10, sigma*2, r"$\bar{r}+$" + r"$2\sigma : {}\ [um]$".format(round(avg+sigma*2,2)), rotation=0, verticalalignment='center')

    # 3*sigma 원 그리기
    circle_3s = plt.Circle((0.0, 0.0), sigma*3, fill=False, color='r', linewidth=1, linestyle='-')
    plt.gca().add_artist(circle_3s)
    text(10, sigma*3, r"$\bar{r}+$" + r"$3\sigma : {}\ [um]$".format(round(avg+sigma*3,2)), rotation=0, verticalalignment='center')

    # 기타
    plt.gca().set_aspect(1)
    plt.xlim(-70, 70)
    plt.ylim(-70, 70)
    plt.xlabel('x [um]')
    plt.ylabel('y [um]')
    plt.grid(True)        
```
